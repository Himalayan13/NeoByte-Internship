```python
import pandas as pd

# Assuming the dataset is a CSV file
csv_file = f"{path}/WA_Fn-UseC_-Telco-Customer-Churn.csv" 
data = pd.read_csv(csv_file)

# Display the first few rows of the dataset
print(data.head())
```

```python
# Converting the categorical numeric column to a numeric column
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
```

```python
# Label encoding the binary columns
binary_columns = ['gender', 'Partner', 'Dependents', 'PhoneService', 'PaperlessBilling', 'Churn']
for col in binary_columns:
    data[col]=data[col].map({'Yes':1, 'No':0, 'Male':1, 'Female':0})
```

```python
multi_category_columns = ['MultipleLines', 'InternetService', 'OnlineSecurity', 
                          'OnlineBackup', 'DeviceProtection', 'TechSupport', 
                          'StreamingTV', 'StreamingMovies', 'Contract', 'PaymentMethod']

data = pd.get_dummies(data, columns=multi_category_columns, drop_first=True)

```

```python
data.info()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Drop 'customerID' since it's not useful for correlation
# data = data.drop(columns=['customerID'])

# Compute correlation matrix
correlation_matrix = data.corr()

# Set figure size
plt.figure(figsize=(12, 8))

# Create heatmap
sns.heatmap(correlation_matrix, 
            annot=True,         # Show correlation values
            fmt=".2f",          # Limit decimal places to 2
            cmap="coolwarm",    # Use a red-blue color map
            linewidths=0.5,     # Add lines between cells
            vmin=-1, vmax=1)    # Ensure colors are properly scaled

# Add title
plt.title("Feature Correlation Heatmap", fontsize=14)

# Show plot
plt.show()
```

```python

# Select numerical columns only
numerical_cols = data.select_dtypes(include=['int64', 'float64'])

# Create a pairplot
sns.pairplot(numerical_cols, diag_kind='kde', corner=True)

# Show the plot
plt.show()
```
